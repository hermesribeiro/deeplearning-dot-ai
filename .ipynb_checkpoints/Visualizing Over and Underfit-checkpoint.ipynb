# Visualizing over and underfit
In this notebook, we present the cases for over and underfit and the impact of lambda parameter in L2 regularization.

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

The main goal of neural networks, as other ML models, is to recover the original function that generated the dataset. In the present case, let's assume our original function is a second degree polinomial.
$$x_2 = ax_1^2+bx_1+c$$
Say, $a=2$, $b=-1$ and $c=1$, the original function would be:

```python
a = 3.
b = -3
c = 1
x1 = np.arange(0,1,0.001)
x2 = a*x1**2+b*x1+c
plt.plot(x1,x2)
plt.ylim(0,1)
plt.xlim(0,1)
plt.show()
```

Now, for a classification problem, let's assume this is a dividing line between two regions. Any point falling below this line has class 0 and above this line has class 1. Let's take some samples.

```python
nsamples = 30
np.random.seed(6)
data = np.random.rand(nsamples,2)
target = (data[:,1] > (a*data[:,0]**2+b*data[:,0]+c)).astype(int)
plt.plot(x1,x2)
plt.scatter(data[(target==1),0],data[(target==1),1])
plt.scatter(data[(target==0),0],data[(target==0),1])
plt.ylim(0,1)
plt.xlim(0,1)
plt.show()
```

Now, let's fit a neural network with arbitrary size.

```python
w1=-0.3
w2=-0.5
w3=-1
b=3
x = np.arange(-10,10,0.01)
z1 =w1*x+b 
z2 =w2*x+b 
z3 =w3*x+b 
y1 = 1/(1+np.exp(-z1))
y2 = 1/(1+np.exp(-z2))
y3 = 1/(1+np.exp(-z3))
plt.plot(x,y1,label="w = 0.3")
plt.plot(x,y2,label="w = 0.5")
plt.plot(x,y3,label="w = 1")
plt.legend()
plt.ylabel('Activation')
plt.xlabel('x')
plt.show()
```

```python
def forward_pass(x,W1,b1):
    z1 = np.dot(W1,x)+b1
    a1 = np.tanh(z1)
    W2 = np.array([1,1,1]).reshape((1,3))
    b2 = 0
    z2 = np.dot(W2,a1)+b2
    a2 = 1/(1+np.exp(-z2))
    return a2
```

```python
x = np.arange(-10,10,0.01).reshape((1,-1))
yhats = np.zeros((4,x.shape[1]))
for i in range(4):
    scale = 10**(-i)
    W1 = np.array([scale,-scale,scale]).reshape((3,1))
    b1 = np.array([-5,0,5]).reshape((3,1))
    yhat = forward_pass(x,W1,b1)
    yhats[i,:] = yhat
plt.plot(x.ravel(),yhats[0,:].ravel(),label=("w = "+str(10**-0)))
plt.plot(x.ravel(),yhats[1,:].ravel(),label=("w = "+str(10**-1)))
plt.plot(x.ravel(),yhats[2,:].ravel(),label=("w = "+str(10**-2)))
plt.plot(x.ravel(),yhats[3,:].ravel(),label=("w = "+str(10**-3)))
plt.legend()
```
